**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
start_time = time.time()

unique_group = list(df['race/ethnicity'].unique())
ethnic_groups = dict.fromkeys(unique_group)

for group in ethnic_groups:
    reading_score_list = []
    for i in range(len(df)):
      if df.iloc[i]['race/ethnicity'] == group:
          reading_score_list.append(df.iloc[i]['reading score'])
    mean_score = sum(reading_score_list)/len(reading_score_list)
    ethnic_groups[group] = mean_score

for i in ethnic_groups:
    print('Reading mean mark based on ' + i + ':', ethnic_groups[i])
print()
print("--- %s seconds ---" % (time.time() - start_time))
print()

start_time = time.time()
unique_degree = list(df['education'].unique())
degrees = dict.fromkeys(unique_degree)
for degree in degrees:
    writing_score_list = []
    for i in range(len(df)):
        if df.iloc[i]['education'] == degree:
            writing_score_list.append(df.iloc[i]['writing score'])
    degrees[degree] = min(writing_score_list)
for i in degrees:
    print('Writing min mark based on ' + i + ':', degrees[i])
print()
print("--- %s seconds ---" % (time.time() - start_time))
print()
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
Math_PassStatus = []
Reading_PassStatus = []
Writing_PassStatus = []

for row in df['math score']:
  if row < passmark:
    Math_PassStatus.append('F')
  else: 
    Math_PassStatus.append('P')

for row in df['reading score']:
  if row < passmark:
    Reading_PassStatus.append('F')
  else: 
    Reading_PassStatus.append('P')

for row in df['writing score']:
  if row < passmark:
    Writing_PassStatus.append('F')
  else: 
    Writing_PassStatus.append('P')

df['Math_PassStatus'] = Math_PassStatus
df['Reading_PassStatus'] = Reading_PassStatus
df['Writing_PassStatus'] = Writing_PassStatus

print('Number of students who passed math exam: ' + str(len(df[df["Math_PassStatus"] == 'P'])))
print('Number of students who failed math exam: ' + str(len(df[df["Math_PassStatus"] == 'F'])))
print()
print('Number of students who passed reading exam: ' + str(len(df[df["Reading_PassStatus"] == 'P'])))
print('Number of students who failed reading exam: ' + str(len(df[df["Reading_PassStatus"] == 'F'])))
print()
print('Number of students who passed writing exam: ' + str(len(df[df["Writing_PassStatus"] == 'P'])))
print('Number of students who failed writing exam: ' + str(len(df[df["Writing_PassStatus"] == 'F'])))
print()
df
```

**Задание 15. Сколько студентов успешно сдали все экзамены?**

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
Overall_PassStatus = []

for i in range(len(df)): 
  math_row = df['math score'][i]
  reading_row = df['reading score'][i]
  writing_row = df['writing score'][i]
  if math_row < passmark or reading_row < passmark or writing_row < passmark:
    Overall_PassStatus.append('F')
  else: 
    Overall_PassStatus.append('P')

df['Overall_PassStatus'] = Overall_PassStatus
print('Number of students who passed all exams: ' + str(len(df[df["Overall_PassStatus"] == 'P'])))
print()
df
```

**Задание 16. Переведем баллы в оценки**

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
def GetGrade(average_mark):
     def_grades = []
     for i in average_mark:
      if i > 90:
          def_grades.append('A')
      elif (80 < i <= 90):
          def_grades.append('B')
      elif (70 < i <= 80):
          def_grades.append('C')
      elif (60 < i <= 70):
          def_grades.append('D')
      elif (50 < i <= 60):
          def_grades.append('E')
      else:
          def_grades.append('F')
     return def_grades

means = df[["math score", "reading score", "writing score"]].mean(axis=1)   
mean_grade = means.tolist()

df['Grade'] = GetGrade(mean_grade)

Grades = ['A', 'B', 'C', 'D', 'E', 'F']
for i in range(len(Grades)):
  print('Number of students with ' + Grades[i] + ': ' + str(len(df[df["Grade"] == Grades[i]])))
  print()
df
```
